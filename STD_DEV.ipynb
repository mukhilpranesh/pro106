```python
import math
import csv

with open ("data.csv", newline="") as f:
  reader=csv.reader(f)
  file_data=list(reader)

data=file_data[0]
```

```python
def mean (data):
  n=len(data)
  total=0
  for x in data:
    total+=int(x)

  mean=total/n

  return mean
```

```python
squared_list=[]
for number in data:
  a=int(number)-mean(data)
  a=a**2
  squared_list.append(a)
```

```python
sum=0
for i in squared_list:
  sum=sum+i
```

```python
result=sum/(len(data)-1)
```

```python
std_dev=math.sqrt(result)

print(std_dev)
```
